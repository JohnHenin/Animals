```python
x= df.iloc[:, [9,10]].values
```

```python
from sklearn.cluster import KMeans
wcss=[]

for i in range(1,35):
    kmeans = KMeans(n_clusters= i, init='k-means++', random_state=0)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)
```

```python
plt.plot(range(1,35), wcss)
plt.title('The Elbow Method')
plt.xlabel('Milch Buffaloes - 2007')
plt.ylabel('Milch Buffaloes - 2012')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
x = df['Milch Cows - Crossbred - % Change']
y = df['Milch Cows - Indigenous - % Change']
z = df['Milch Buffaloes - % Change']
x_train, x_test, y_train, y_test,z_train,z_test= train_test_split(x, y,z, train_size = 0.75, shuffle = True)
from sklearn.linear_model import LinearRegression
print(x_train.shape)
print(x_test.shape)
```

```python
import numpy as np
x_train=np.array(x_train).reshape(-1,1)
y_train=np.array(y_train).reshape(-1,1)

model=LinearRegression()
model. fit(x_train,y_train)

print (model. coef_)
print(model.intercept_)
```

```python
y_test=np.array(y_test).reshape(-1,1)
y_pred=model.predict(y_test)
print(y_pred)
```

```python
z_test=np.array(z_test).reshape(-1,1)
z_pred=model.predict(y_test)
print(z_pred)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
import math

print('Mean absolute error:', mean_absolute_error(y_test, y_pred))

print("Mean squared error:", mean_squared_error(y_test, y_pred))
print('variance score:',r2_score(y_test, y_pred))

print('Root Mean Squared Error:',math.sqrt(mean_squared_error(y_test, y_pred)))
```
